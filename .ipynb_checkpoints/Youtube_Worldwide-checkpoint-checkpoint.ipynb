The country_tags_dic_sort function takes one paramater a given country and removes all the tags of that given country and puts it into a dictionary with the tags as the key and the count of occurances of the tags as the value.
Processing this data made me relize that the tags might have little effect on the trendingness of the videos. This is becasue the leading tag in every country was the fact the the video had no tags attatched to it what so ever. Which leads me to belive that the tags are used more to categorize the videos which in turn would lead to differnt trends.
You can run this function with any of the given 11 given countries to return a dictionary of tags and tag_frequences.

```python
def country_tags_dic_sort(given_country):
    country_tags_sublist = result.loc[result['country'] == given_country, 'tags'].tolist()
    for i in range(len(country_tags_sublist)):
        country_tags_sublist[i] = country_tags_sublist[i].split('|')
    country_tags_list = list(itertools.chain.from_iterable(country_tags_sublist))
    country_tags_dic = {}
    for j in country_tags_list:
        country_tags_dic[j] = country_tags_dic.get(j, 0) + 1
    country_tags_dic_sorted = dict(sorted(country_tags_dic.items(), key = lambda kv:(kv[1], kv[0]), reverse = True))
    del country_tags_dic_sorted['[None]']
    return country_tags_dic_sorted
#country_tags_dic_sort("US")
```

The top_country_tags function takes two paramaters a given country and the number of top tags you would like to recive from the given countries. This function returns a list of tuples. 

```python
def top_country_tags(country, top):
    return more_itertools.take(top, country_tags_dic_sort(country).items())
#print(top_10_country_tags("US"))
```

The display_top_country_tags function takes no parameters and is called by the command line argument -s, --show. This function displayes the top 20 tags of each country removing the '[None]' tag as its significance has already been addressed in the data. The data is shown as a histogram with the tags on the x-axis and the frequencey on the y-axis. 
You can uncomment the translator to get a better data set. However google might block your ip becasue I am making an extream amount of calls do their api and ran out of time to do it efficeintly. In displaying the data this way I saw corrilation between many english and spanish speaking countryings with tags such as "funny", "futbol", "humor", and "comedy". With the translator we can see much more correlation with RU, KR, and JP in realation to the other countries. And even without the translator we still the "2020" and "vlog" in almost all countries. So as differnt countries have different languages and cultures there is still a correlation between the top tags used in each different country. 

```python
def display_top_country_tags():
    country_list = result['country'].unique()
#    translator = google_translator()
    for i in country_list:
        list1 = top_country_tags(i,20)
        word, frequency = zip(*list1)
        word = list(word)
#        for j in range(len(word)):
#            word[j] = translator.translate(word[j])
        indices = np.arange(len(list1))
        plt.rcParams["font.serif"] = "UTF-8"
        plt.bar(indices, frequency, color = 'r')
        plt.xticks(indices, word, rotation ='vertical')
        plt.title("Frequency of Top 25 Tags-%s"%i)
        plt.tight_layout()
        plt.show()
        print(word)
display_top_country_tags()
```

The display_country_tagpercentage function takes no parameters and is called by the command line argument -s, --show. This function displays the percentage of usage of the top 10 tags for each country. These pie charts show use how often the top words are used in relation to the other top 10 words. What I found here was that none of the top tags where unevenly distributed. There was no one tag that was used miles above the rest. I tought that there would be one tag used much more frequently in each individual country. But I was wrong there is a more even distribution of the top tags. 

```python
def display_country_tagpercentage():
    country_list=result['country'].unique()
#    translator = google_translator()
    for i in country_list:
        list1 = top_country_tags(i, 10)
        labels, size = zip(*list1)
        labels = list(labels)
#        for j in range(len(labels)):
#            labels[j] = translator.translate(labels[j])
        fig1, ax1 = plt.subplots() 
        ax1.pie(size, labels=labels, autopct='%1.1f%%', shadow=True)
        ax1.set(aspect = 'equal', title = "Top Tags - %s"%i)
        plt.show()
display_country_tagpercentage()
```

The cross_referance_tags function takes three paramaters two countries and an integer representing the number of top tags you would like to cross reference per country. Again this function works alot better if I could get the translator to work properly; you are welcolme to uncoment it and try it out. It works on my system 1/10 times. This function cross referances the tags of two countries and returns a list of all the elements that appear in both countries top tags. It is very interesting to look at the intersection between differnt countries. 

```python
def cross_referance_tags(country1, country2, number_of_top_tags=50):
#    translator = google_translator()
    country1_dic = {}
    for k,v in top_country_tags(country1, number_of_top_tags):
#        k = translator.translate(k)
        country1_dic.setdefault(k, []).append(v)
    country2_dic = {}
    for k,v in top_country_tags(country2, number_of_top_tags):
#        k = translator.translate(k)
        country2_dic.setdefault(k, []).append(v)
    intersect = []
    for item in country1_dic.keys():
        if item in country2_dic.keys():
            intersect.append(item)
    return intersect
print(cross_referance_tags('MX', 'US', 10))
print(cross_referance_tags('MX', 'US', 1000))
print(cross_referance_tags('US', 'CA'))
#the top 3 in the US and CA are exactly the same "Funny", "comedy" and "2020".
print(cross_referance_tags('US', 'CA', 3))
```

This is where it all comes together. The tag_cross_referance takes two paramaters a starting country it can be any country just to start the cross referance and the number of tags you would like to cross referance the higher the number the more tags will be cross referanced and you will likly have more output. I had to remove a few data sets that had most of their tags in a differnt languages. None the less I found that between the mainly english and spanish speaking countries, [Brazil, Canada, Denmark, France, Great Britan, India, Mexico, United States] There was a tremendous amount of corrilation between the tags used. Leading me to belive that many of the top videos across these contries have similar types of videos, as far as content and categories. When I was able to run the program with the translator I found even more corrilation between all of the 11 countries in the data set. Tags such as 'vlog', 'youtube', '2020', 'anime', 'funny', 'comedy', 'soccer' are just a few of the words that showed up in all of the translated data sets. It was very interesting to see the similarities between the different cultues. 

```python
def tag_cross_referance(country1, top=50):
    country_list = result['country'].unique()
    index =[6,7,9]
    country_list = np.delete(country_list, index)
    temp = []
    intersect = []
    for i in country_list:
        if i != country1:
            for item in cross_referance_tags(country1, i, top):
                temp.append(item)
    for j in temp:
        if temp.count(j) >= 7:
            if j not in intersect:
                intersect.append(j)
    return intersect
tag_cross_referance("BR",100) 
```

The argparser has 3 arguments -s to show the pie and/or histogram charts for the tags data. -t [integer] to show the list of cross referanced tags between all the countries looking at the given top number of tags. 

```python
parser = argparse.ArgumentParser(description="Analyze Top trending YouTube data sets")
parser.add_argument('Command', metavar='<command>', type=str, choices = ['print'], help='command to execute: only choice is to print')
parser.add_argument('-s','--show', dest = 'show', type=str, action = 'store', metavar = '<show>', choices=['Pie', 'Histogram'], help = 'command to execute: display_country_tagpercentage, display_top_country_tags')
parser.add_argument('-t', '--tags', dest = 'tags', type=int, action = 'store', metavar = '<tags>', choices=["An integer 0-10,000"], help = 'command to execute: tag_cross_referance')
args = parser.parse_args()
```

```python
if args.show is not none:
    if args.show == 'Pie':
        display_country_tagpercentage()
    elif args.show == 'Histogram':
        display_top_country_tags()
    else:
        Print("Invalid choice for --show")
        
if args.tags is not none:
    if (0 < args.tags <= 10000):  
        tag_cross_referance("BR",args.tags)
    else:
        print("Invalid choice for --tags")
```
